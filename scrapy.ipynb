```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
page=requests.get(url)
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
name=soup.findAll(attrs={'class':'_4rR01T'})
print(name)
```

```python
kcname=[]
for i in name:
    print(i.text)
    kcname.append(i.text)
```

```python
print(kcname)
```

```python
price=soup.findAll(attrs={'class':'_30jeq3 _1_WHN1'})
kcprice=[]
for i in price:
    print(i.text.replace('','Rs .'))
    kcprice.append(i.text.replace('','Rs .'))
```

```python
print(kcprice)
```

```python
discount=soup.findAll(attrs={'class':'_3Ay6Sb'})
kcdiscount=[]
for i in discount:
    print(i.text)
    kcdiscount.append(i.text)
```

```python
print(kcdiscount)
```

```python
imgurl=soup.findAll(attrs={'class':'_396cs4'})
kcimgurl=[]
for i in imgurl:
    print(i.get('src'))
    kcimgurl.append(i.get('src'))
```

```python
print(kcimgurl)
```

```python
data=[kcname,kcprice,kcdiscount,kcimgurl]
```

```python
print(data)
```

```python
df=pd.DataFrame(data,columns=['item 1','item 2','item 3','item 4','item 5','item 6','item 7','item 8','item 9','item 10','item 11','item 12','item 13','item 14','item 15','item 16','item 17','item 18','item 19','item 20','item 21','item 22','item 23','item 24','item 25','item 26','item 27','item 28','item 29'])
```

```python
print(df)
```

```python
df.to_csv('pagedataa.csv')
```
